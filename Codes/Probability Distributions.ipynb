# Distributions

In probability theory and statistics, a probability distribution is the mathematical function that gives the probabilities of occurrence of different possible outcomes for an experiment.[2][3] It is a mathematical description of a random phenomenon in terms of its sample space and the probabilities of events (subsets of the sample space).[4]

## 1. Continous Distributions

A continuous probability distribution is a probability distribution whose support is an uncountable set, such as an interval in the real line.[5] They are uniquely characterized by a cumulative distribution function that can be used to calculate the probability for each subset of the support.

<h3>1.5 F Distribution</h3>

[Python Documentation](https://numpy.org/doc/stable/reference/random/generated/numpy.random.f.html)

```python
# dfnum: between group degrees of freedom
# dfden: within groups degrees of freedom
plt.figure(figsize=(20,10))

# F Distribution with dfnum=1, dfden=1
plt.subplot(2,2,1)
s = np.random.f(dfnum=1, dfden=1, size=10000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins=100, density=True)
plt.title("F Distribution (dfnum=1, dfden=1)", fontsize=16)

# F Distribution with dfnum=1, dfden=10
plt.subplot(2,2,2)
s = np.random.f(dfnum=1, dfden=10, size=10000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins=100, density=True)
plt.title("F Distribution (dfnum=1, dfden=10)", fontsize=16)


# F Distribution with dfnum=10, dfden=1
plt.subplot(2,2,3)
s = np.random.f(dfnum=10, dfden=1, size=10000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins=100, density=True)
plt.title("F Distribution (dfnum=10, dfden=1)", fontsize=16)

# F Distribution with dfnum=10, dfden=10
plt.subplot(2,2,4)
s = np.random.f(dfnum=10, dfden=10, size=10000)
s = s[np.where(abs(s) < 5)]
h = plt.hist(s, bins=100, density=True)
plt.title("F Distribution (dfnum=10, dfden=10)", fontsize=16)
plt.savefig("Fdistribution.png")
plt.show()
```

<h3>1.10 Birnbaum-Saunders (Fatigue Life) Distribution</h3>

[Python Documentation](https://docs.scipy.org/doc/scipy/reference/generated/scipy.stats.fatiguelife.html)

```python
from scipy.stats import fatiguelife
plt.figure(figsize=(20,10))

# fatiguelife Distribution with gamma = 0.5
plt.subplot(2,2,1)
x = np.linspace(fatiguelife.ppf(0.01, c=0.5), fatiguelife.ppf(0.99, c=0.5), 100)
plt.plot(x, fatiguelife.pdf(x, c=0.5), 'b-', lw=2, alpha=0.6, label='fatiguelife pdf')
plt.title("fatiguelife Distribution (gamma=0.5)", fontsize=16)

# fatiguelife Distribution with gamma = 1
plt.subplot(2,2,2)
x = np.linspace(fatiguelife.ppf(0.01, c=1), fatiguelife.ppf(0.99, c=1), 100)
plt.plot(x, fatiguelife.pdf(x, c=1), 'b-', lw=2, alpha=0.6, label='fatiguelife pdf')
plt.title("fatiguelife Distribution (gamma=1)", fontsize=16)


# fatiguelife Distribution with gamma = 2
plt.subplot(2,2,3)
x = np.linspace(fatiguelife.ppf(0.01, c=2), fatiguelife.ppf(0.99, c=2), 100)
plt.plot(x, fatiguelife.pdf(x, c=2), 'b-', lw=2, alpha=0.6, label='fatiguelife pdf')
plt.title("fatiguelife Distribution (gamma=2)", fontsize=16)

# fatiguelife Distribution with gamma = 5
plt.subplot(2,2,4)
x = np.linspace(fatiguelife.ppf(0.01, c=5), fatiguelife.ppf(0.99, c=5), 100)
plt.plot(x, fatiguelife.pdf(x, c=5), 'b-', lw=2, alpha=0.6, label='fatiguelife pdf')
plt.title("fatiguelife Distribution (gamma=5)", fontsize=16)
plt.savefig("fatiguelifedistribution.png")
plt.show()
```

## 2. Discrete Distributions

A discrete probability distribution is the probability distribution of a random variable that can take on only a countable number of values.[6] In the case where the range of values is countably infinite, these values have to decline to zero fast enough for the probabilities to add up to 1

__Examples of Discrete distributions:__

2.1 [Binomial Distribution](#DD1)<br>
2.2 [Poisson Distribution](#DD2)

<a id="DD1"></a>
<h3>2.1 Binomial Distribution</h3>

[Python Documentation](https://numpy.org/doc/stable/reference/random/generated/numpy.random.binomial.html)

```python
plt.figure(figsize=(20,10))

# Binomial Distribution with (n=100, p=0.1)
plt.subplot(2,2,1)
n, p = 100, .1  # number of trials, probability of each trial
s = np.random.binomial(n, p, 10000)
h = plt.hist(s, bins=20, density=True)
plt.title("Binomial Distribution (n=100, p=0.1)", fontsize=16)

# Binomial Distribution with (n=100, p=0.25)
plt.subplot(2,2,2)
n, p = 100, .25  # number of trials, probability of each trial
s = np.random.binomial(n, p, 10000)
h = plt.hist(s, bins=20, density=True)
plt.title("Binomial Distribution (n=100, p=0.25)", fontsize=16)


# Binomial Distribution with (n=100, p=0.5)
plt.subplot(2,2,3)
n, p = 100, .5  # number of trials, probability of each trial
s = np.random.binomial(n, p, 10000)
h = plt.hist(s, bins=20, density=True)
plt.title("Binomial Distribution (n=100, p=0.5)", fontsize=16)

# Binomial Distribution with (n=100, p=0.75)
plt.subplot(2,2,4)
n, p = 100, .75  # number of trials, probability of each trial
s = np.random.binomial(n, p, 10000)
h = plt.hist(s, bins=20, density=True)
plt.title("Binomial Distribution (n=100, p=0.75)", fontsize=16)
plt.savefig("Binomialdistribution.png")
plt.show()
```

<h3>2.2 Poisson Distribution</h3>

[Python Documentation](https://numpy.org/doc/stable/reference/random/generated/numpy.random.poisson.html)

```python
plt.figure(figsize=(20,10))

# Poisson Distribution with (lambda = 5)
plt.subplot(2,2,1)
s = np.random.poisson(lam=5, size=100000)
h = plt.hist(s, bins=20, density=True)
plt.title("Poisson Distribution (lambda = 5)", fontsize=16)

# Poisson Distribution with (lambda = 15)
plt.subplot(2,2,2)
s = np.random.poisson(lam=15, size=100000)
h = plt.hist(s, bins=20, density=True)
plt.title("Poisson Distribution (lambda = 15)", fontsize=16)


# Poisson Distribution with (lambda = 25)
plt.subplot(2,2,3)
s = np.random.poisson(lam=25, size=100000)
h = plt.hist(s, bins=20, density=True)
plt.title("Poisson Distribution (lambda = 25)", fontsize=16)

# Poisson Distribution with (lambda = 35)
plt.subplot(2,2,4)
s = np.random.poisson(lam=5, size=100000)
h = plt.hist(s, bins=20, density=True)
plt.title("Poisson Distribution (lambda = 35)", fontsize=16)
plt.savefig("Poissondistribution.png")
plt.show()
```
